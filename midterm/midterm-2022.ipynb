## Step 1: Calculate PRIEST Clinical Severity Score

This scoring algorithm can be found [here on the MDCalc website](https://www.mdcalc.com/priest-covid-19-clinical-severity-score#evidence).  

1. You will need to write a function called **priest()** with the following input parameters.  
 * Sex
 * Age in years
 * Respiratory rate in breaths per minute
 * Oxygen saturation as a percent between 0 and 1
 * Heart rate in beats per minute
 * Systolic BP in mmHg
 * Temperature in degrees C
 * Alertness as a string description
 * Inspired Oxygen as as string description
 * Performance Status as a string description
2. The function will need to follow the algorithm provided on the MDCalc website to compute a risk percentage that should be returned as a numeric value between 0 and 1.
3. Be sure to use docstring documentation and at least three built-in docstring test cases.
4. Assume that the input values that are strings could be any combination of upper or lower case. For example: 'male', 'Male', 'MALE', 'MalE' should all be interpretted by your code as male.
5. If any of the inputs are invalid (for example a sex value that is not recognizable as male or female) your code should raise a ValueException that includes a message with the invalid input and which parameter that input was provided as.

NOTES:
1. In the final step there is a rule to convert form raw score to percentile.  In that conversion, 17-25 maps to 59-88% and â‰¥26 maps to >99%.  For our purposes, we want these to be specific % number outputs.  Use the following rule:
 * If score is between 17 and 25, percentile should be 0.59
 * If score is greater than or equal to 26, percentile should be 0.99


````python
import sys 
import os 
def priest(sex, age, resp_rate, oxy_sat, heart_rate, sys_bp, temp, alert, insp_oxy, per_stat):
    """(str, int, int, int, int, int, float, str, str, str) -> float
    This function computes a risk percentage based on the algorithm from the MDCalc website
    that should be returned as a numeric value between 0 and 1. 
    >>> priest("Male",90,6,92,41,91,35.1,"Alert","Air","Unrestricted normal activity")
    0.47
    >>> priest("FeMale",17,10,92,135,91,35.1,"confused or not alert","supplemental oxygen","Limited self-care" )
    0.59
    >>> priest("male",90,35,85,135,250,30,"confused or not alert","supplemental oxygen","bed/chair bound, no self-care")
    0.99
    """
    points = 0
# Sex    
    try:
        if (sex.lower() in ['male']):
            points = points + 1
        elif (sex.lower() in ['female']):
            points = points + 0
        else:
            Ex = ValueError()
            Ex.strerror = "Invalid values for sex : " + str(sex)
            raise Ex
    except ValueError as e:
            print("ValueError Exception!", e.strerror)
# Age
    try:
        if age in range(16,50):
            points = points + 0
        elif age in range(50,66):
            points = points + 2
        elif age in range(66,81):
            points = points + 3
        elif age > 80:
            points = points + 4
        else:
            Ex = ValueError()
            Ex.strerror = "Invalid values for Age : " + str(age)
            raise Ex
    except ValueError as e:
            print("ValueError Exception!", e.strerror)
            
# Respiratory Rate
    try:
        if resp_rate < 9:
            points = points + 3
        elif resp_rate in range(9,12):
            points = points + 1
        elif resp_rate in range(12,21):
            points = points + 0
        elif resp_rate in range(21,25):
            points = points + 2
        elif resp_rate > 24:
            points = points + 3
        else:
            Ex = ValueError()
            Ex.strerror = "Invalid values for Respiratory Rate : " + str(resp_rate)
            raise Ex
    except ValueError as e:
            print("ValueError Exception!", e.strerror)

#Oxygen Saturation 
    try: 
        if oxy_sat > 95: 
            points = points + 0 
        elif oxy_sat in range(94,96): 
            points = points + 1
        elif oxy_sat in range(92,94):
            points = points + 2 
        elif oxy_sat < 92: 
            points = points + 3 
        else:
            Ex = ValueError()
            Ex.strerror = "Invalid values for oxygen saturation : " + str(oxy_sat) 
            raise Ex
    except ValueError as e:
            print("ValueError Exception!", e.strerror)

#Heart Rate 
    try: 
        if heart_rate < 41:
            points = points + 3 
        elif heart_rate in range(41,51):
            points = points + 1 
        elif heart_rate in range(51,91): 
            points = points + 0 
        elif heart_rate in range(91,111):
            points = points + 1 
        elif heart_rate in range(111,131):
            points = points + 2
        elif heart_rate > 130: 
            points = points + 3
        else:
            Ex = ValueError()
            Ex.strerror = "Invalid values for heart rate : " + str(heart_rate)
            raise Ex
    except ValueError as e:
            print("ValueError Exception!", e.strerror)

#Systolic BP, mmHG 
    try: 
        if sys_bp < 91: 
            points = points + 3 
        elif sys_bp in range(91,101):
            points = points + 2 
        elif sys_bp in range(101,111):
            points = points + 1
        elif sys_bp in range(111,220):
            points = points + 0
        elif sys_bp > 219: 
            points = points + 3 
        else:
            Ex = ValueError()
            Ex.strerror = "Invalid values for systolic BP : " + str(sys_bp) 
            raise Ex
    except ValueError as e:
            print("ValueError Exception!", e.strerror)

#Temperature 
    try: 
        if temp < 35.1:
            points = points + 3
        elif temp >= 35.1 and temp <= 36.0 :
            points = points + 1 
        elif temp >= 36.1 and temp <= 38.0:
            points = points + 0 
        elif temp >= 38.1 and temp <= 39.0: 
            points = points + 1 
        elif temp > 39.0: 
            points = points + 2 
        else:
            Ex = ValueError()
            Ex.strerror = "Invalid values for temperature : " + str(temp) 
            raise Ex
    except ValueError as e:
            print("ValueError Exception!", e.strerror)    
            
#Alertness 
    try:
        if (alert.lower() in ['alert']):
            points = points + 0 
        elif (alert.lower() in ['confused or not alert']):
            points = points + 3 
        else:
            Ex = ValueError()
            Ex.strerror = "Invalid values for Age : " + str(alert)
            raise Ex
    except ValueError as e:
            print("ValueError Exception!", e.strerror) 

#Inspired Oxygen 
    try: 
        if (insp_oxy.lower() in ['air']):
            points = points + 0 
        elif(insp_oxy.lower() in ['supplemental oxygen']):
            points = points + 2 
        else:
            Ex = ValueError()
            Ex.strerror = "Invalid values for Inspired Oxygen : " + str(insp_oxy)
            raise Ex
    except ValueError as e:
            print("ValueError Exception!", e.strerror) 

#Performance Status 
    try: 
        if (per_stat.lower() in ['unrestricted normal activity']):
            points = points + 0  
        elif (per_stat.lower() in ['limited strenuous activity, can do light activity']):
            points = points + 1 
        elif (per_stat.lower() in ['limited activity, can self-care']):
            points = points + 2 
        elif (per_stat.lower() in ['limited self-care']):
            points = points + 3 
        elif (per_stat.lower() in ['bed/chair bound, no self-care']):
            points = points + 4 
        else:
            Ex = ValueError()
            Ex.strerror = "Invalid values for Performance Status : " + str(per_stat)
            raise Ex
    except ValueError as e:
            print("ValueError Exception!", e.strerror) 

            
         
 #Calculate Risk Percentage
    riskpercentage = 0

    if points in range(0,2):
        riskpercentage = 1
    elif points in range(2,4):
        riskpercentage = 2
    elif points == 4:
        riskpercentage = 3
    elif points == 5:
        riskpercentage = 9
    elif points == 6:
        riskpercentage = 15
    elif points == 7:
        riskpercentage = 18
    elif points == 8:
        riskpercentage = 22
    elif points == 9:
        riskpercentage = 26
    elif points == 10:
        riskpercentage = 29
    elif points == 11:
        riskpercentage = 34
    elif points == 12:
        riskpercentage = 38
    elif points == 13:
        riskpercentage = 46
    elif points == 14:
        riskpercentage = 47
    elif points == 15:
        riskpercentage = 49
    elif points == 16:
        riskpercentage = 55
    elif points in range(17,26):
        riskpercentage = 59
    elif points >= 26:
        riskpercentage = 99

            
    return riskpercentage/100
        
    
````

## Part 2: Find a hospital

The next thing we have to do is figure out where to send this particular patient.  The guidelines on where to send a patient are based on their age (pediatric, adult, geriatric), sex, and risk percentage.  Luckily, you don't have to implement these rules. I already have. All you have to do is use a REST web service that I've created for you.

You'll want to use Python to make a call to my REST web service similar to the example URL below. The first part of the URL will be the same for everyone and every request that you make. What you will need to modify for each of your requests is the information after the question mark.

```
https://oumdj6oci2.execute-api.us-east-1.amazonaws.com/prd/?age=40&sex=male&risk_pct=0.1
```

The example above asks my web service where a 40-year old male with a risk percetage of 10% should go.  What the web service will return back is a JSON string containing the information you need.  That JSON will look like this:

```json
{
  "age": "40",
  "sex": "male",
  "risk": "0.1",
  "hospital": "Southwest Hospital and Medical Center"
}
```

1. Your job is to write a function called **find_hospital()** that takes age, sex, and risk as parameters.
2. Your function should call this REST web service
3. Then your function will need to interpret the JSON it gets and return just the name of the hospital
4. If anything fails, return None without raising any exceptions
5. Include a good docstring with at least five test cases.


````python
import requests 
def find_hospital(age, sex, riskpercentage):
    """(int,str,float) -> str 
    This function takes age, sex, and risk as parameters and calls a REST web service that 
    interprets the JSON and returns just the name of a hospital 
    >>> find_hospital('40','male','0.59')
    'Emory Dunwoody Medical Center'
    >>> find_hospital('10','female','0.30')
    'Childrens Healthcare of Atlanta at Scottish Rite'
    >>> find_hospital('80','male','0.99')
    'Wesley Woods Geriatric Hospital'
    >>> find_hospital('35','female','0.40')
    'Emory Dunwoody Medical Center'
    >>> find_hospital('50','male',0.20)
    'Southwest Hospital and Medical Center'
      
    """
    try: 
        response = requests.get('https://oumdj6oci2.execute-api.us-east-1.amazonaws.com/prd/?age='+str(age)+'&sex='+str(sex)+'&risk_pct='+str(riskpercentage)+'')
        data = response.json()
    except ValueError as e:
        return ("None")
    return data['hospital']
    
````

## Part 3: Get the address for that hospital from a webpage

Great! Now we have code to tell us which hospital to send someone to... but we don't know where that hospital is. The next function we need to create is one that looks up the address of that hospital.  All of these hospitals are in Atlanta, Georgia.  We're going to use the list on this webpage to lookup the address for that hospital, based on its name.

https://www.officialusa.com/stateguides/health/hospitals/georgia.html

1. You need to create a function called **get_address()** that takes hospital name as a parameter and searches the data on the webpage above to find the addres for that hospital.
2. I've said that all the hospitals are in Atlanta, but this webpage has hospitals from all of Georgia.  So, make sure you verify that the row of data you're using is in Atlanta, just in case there are hospitals with the same name in different cities.
3. Your code will have to parse the HTML on the webpage and turn that into some kind of data structure that you can search through to find the right hospital.
4. If you do find more than one matching hospital in Atlanta with the same name, you should raise a KeyError.
5. If the hospital name isn't found, the function should raise a KeyError.
6. Be sure to use good docstring documentation and includes at least 3 test cases.

````python
from bs4 import BeautifulSoup
import requests

def get_address(hospitalname): 
    """(str) -> str
    This function takes hospital name as a parameter and searched the url to return
    the address for the hospital 
    >>> get_address('SELECT SPECIALTY HOSPITAL - ATLANTA')
    '550 PEACHTREE STREET, NE'
    >>> get_address('PRIDE MEDICAL')
    '3280 HOWELL MILL ROAD NW'
    >>> get_address('GRADY MEMORIAL HOSPITAL')
    '80 JESSE HILL, JR DRIVE SE'
    """
    headers = { "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 11_2_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36" }
    response = requests.get('https://www.officialusa.com/stateguides/health/hospitals/georgia.html', headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    
    table = soup.find('table', attrs={"id":"myTable"})
    rows = table.find_all('tr')
    hospital_address = {}

    for row in rows: 
        colno = 0 
        atlanta = 'false'
        hospital_name = ""
        address = "" 
        for col in row.find_all('td'):
            if colno == 0 and col.text.lower() == 'atlanta':
                atlanta = 'true'
            if colno == 1:
                hospital_name = col.text
            if colno == 3: 
                address = col.text 
            if colno in [1,3]:
                    if atlanta == 'true':
                        hospital_address[hospital_name] = address

            colno = colno + 1 
    
    
            
    return hospital_address.get(hospitalname)

    
    
    

    
    
    
````

Your Python program should use your code above to process all of these rows, determine the hospital and address, and return a list whose items are a dictionary like `{ patient_number: [sex, age, breath, o2sat, heart, systolic, temp, alertness, inspired, status, hospital, address]}`

**NOTE** that when running your code for all the 100 records in the `people.psv` file, it may take a few minutes to complete.  You're making multiple calls to the internet for each record, so that can take a little while.


````python
import pandas as pd
import os
import json 
def process_people(url): 
    """(url) -> .json file 
    This function joins lists to return a dictionary with various elements of a patient """
    df = pd.read_csv(url, sep= '|')
    patientlist = df['patient'].tolist() 
    sexlist = df['sex'].tolist()
    agelist = df['age'].tolist()
    breathlist = df['breath'].tolist()
    o2satlist = df['o2 sat'].tolist()
    heartrtlist = df['heart'].tolist()
    sysbplist = df['systolic bp'].tolist()
    templist = df['temp'].tolist()
    alertlist = df['alertness'].tolist()
    inspiredlist = df['inspired'].tolist()
    statuslist = df['status'].tolist()
    hospitalnamelist = [] 
    hospitaladdresslist = [] 
    riskpercentagelist = []
    people_dict = {} 
    for index in range(len(patientlist)):
        riskpercentage = priest(str(sexlist[index]), int(agelist[index]), int(breathlist[index]), int(o2satlist[index]* 100), int(heartrtlist[index]), int(sysbplist[index]), float(templist[index]), str(alertlist[index]), str(inspiredlist[index]), str(statuslist[index]))
        hospital_name = find_hospital(int(agelist[index]), str(sexlist[index]), float(riskpercentage))
        hospital_address = get_address(str(hospital_name.upper()))
        hospitalnamelist.append(hospital_name)
        hospitaladdresslist.append(hospital_address)
        riskpercentagelist.append(riskpercentage)

       
    zipped = zip(sexlist,agelist,breathlist,o2satlist,heartrtlist,sysbplist,templist,alertlist,inspiredlist,statuslist,riskpercentagelist,hospitalnamelist,hospitaladdresslist)
    people_dict = dict(zip(patientlist, list(zipped)))
    return people_dict

print(process_people('/data/people.psv'))

#os.getcwd()
    

    #print(process_people('/data/people.psv'))

with open("ashleys_patient_list.json","w") as outfile:
    json.dump(process_people('/data/people.psv'), outfile)

    
````
